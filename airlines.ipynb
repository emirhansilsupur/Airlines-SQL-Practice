### 1. What is the distribution of different aircraft models in the fleet?

```python
query1 = '''
SELECT aircraft_code,model
FROM aircrafts_data;
'''
run_query(query1)
```

### 2. What are the range capabilities of the different aircraft models?

```python
query2 = '''
WITH ModelExtraction AS (
    SELECT
        json_extract(model, '$.en') AS model_name,
        range
    FROM
        aircrafts_data
)

SELECT
    ModelExtraction.model_name,
    ModelExtraction.range
FROM
    ModelExtraction
ORDER BY
    range DESC;
'''
run_query(query2)
```

### 3. Which are the top 10 flight routes by the number of flights?

```python
query3 = '''
SELECT departure_airport, arrival_airport, COUNT(*) AS flight_count
FROM flights
GROUP BY departure_airport,arrival_airport
ORDER BY flight_count DESC
LIMIT 10;
'''
run_query(query3)
```

### 4. What are the geographic locations and timezones of these airports?

```python
query4 = '''
WITH top_departure_airports AS (
    SELECT departure_airport AS airport_code, COUNT(*) AS airport_count
    FROM flights
    GROUP BY departure_airport
    ORDER BY airport_count DESC
    LIMIT 10
),
top_arrival_airports AS (
    SELECT arrival_airport AS airport_code, COUNT(*) AS airport_count
    FROM flights
    GROUP BY arrival_airport
    ORDER BY airport_count DESC
    LIMIT 10
),
combined_airports AS (
    SELECT airport_code
    FROM top_departure_airports
    UNION
    SELECT airport_code
    FROM top_arrival_airports
),
AirportDataEn AS (
  SELECT airport_code,json_extract(airport_name, '$.en') AS airport_name_en, json_extract(city, '$.en') AS city_en, timezone FROM airports_data
)
SELECT airport_code, airport_name_en, city_en, timezone
FROM AirportDataEn
WHERE airport_code IN (SELECT airport_code FROM combined_airports)
ORDER BY airport_code;
'''

run_query(query4)
```

### 5. What are the total flights per aircraft model?

```python
query5 = """
WITH ModelExtraction AS (
    SELECT 
        aircraft_code,
        json_extract(model, '$.en') AS model_name
    FROM 
        aircrafts_data
)

SELECT f.aircraft_code,mdl.model_name, COUNT(*) AS total_flights
FROM flights f
JOIN ModelExtraction mdl ON f.aircraft_code = mdl.aircraft_code
GROUP BY 
    f.aircraft_code, mdl.model_name
ORDER BY 
    total_flights DESC;
"""
run_query(query5)
```

### 6. How do boarding patterns vary across different flights and routes?

```python
query6 = """
SELECT bp.flight_id, f.flight_no, departure_airport,arrival_airport, COUNT(bp.flight_id) AS boarding_count
FROM boarding_passes bp
INNER JOIN flights f ON bp.flight_id = f.flight_id
GROUP BY bp.flight_id
ORDER BY boarding_count DESC
LIMIT 10;
"""
run_query(query6)
```

### 7. What Are the Most Frequently Occupied Seats by Aircraft and Economy Class?

```python
query7 = '''
SELECT
    f.aircraft_code,
    bp.seat_no,
    s.fare_conditions,
    COUNT(bp.seat_no) AS seat_count
FROM
    boarding_passes bp
JOIN
    flights f ON bp.flight_id = f.flight_id
JOIN
    seats s ON f.aircraft_code = s.aircraft_code AND bp.seat_no = s.seat_no
WHERE
    s.fare_conditions = 'Economy'
GROUP BY
    f.aircraft_code, bp.seat_no, s.fare_conditions
ORDER BY seat_count DESC
LIMIT 5;
'''
run_query(query7)
```

### 8. What Are the Most Frequently Occupied Seats by Aircraft and Business Class?

### 10. What Are the Least Frequently Occupied Seats by Aircraft and Fare Class?

```python
query10 = '''
SELECT
    f.aircraft_code,
    bp.seat_no,
    s.fare_conditions,
    COUNT(bp.seat_no) AS seat_count
FROM
    boarding_passes bp
JOIN
    flights f ON bp.flight_id = f.flight_id
JOIN
    seats s ON f.aircraft_code = s.aircraft_code AND bp.seat_no = s.seat_no
GROUP BY
    f.aircraft_code, bp.seat_no, s.fare_conditions
ORDER BY seat_count ASC
LIMIT 5;
'''
run_query(query10)
```

###  11. What are the on-time departure and arrival rates for flights?

```python
query11 = r'''
SELECT COUNT(*) AS total_flights,
       ROUND(SUM(CASE WHEN actual_departure <= scheduled_departure THEN 1 ELSE 0 END) * 100.0 / COUNT(*),2) AS on_time_departure_rate,
       ROUND(SUM(CASE WHEN actual_arrival <= scheduled_arrival THEN 1 ELSE 0 END) * 100.0 / COUNT(*),2) AS on_time_arrival_rate
FROM flights
WHERE actual_departure != '\N' AND actual_arrival != '\N';
'''
run_query(query11)
```

### 13. What is the distribution of different fare conditions across flights?

```python
query13 = '''
SELECT fare_conditions, COUNT(*) AS fare_count
FROM ticket_flights
GROUP BY fare_conditions
ORDER BY fare_count DESC;
'''
run_query(query13)
```

### 14. What are the revenue contributions of different fare conditions and routes?

```python
query14 = '''
SELECT fare_conditions, SUM(amount) AS total_revenue, COUNT(*) AS ticket_count, SUM(amount) / COUNT(*) AS average_revenue_per_ticket
FROM ticket_flights
GROUP BY fare_conditions
ORDER BY total_revenue DESC;
'''
run_query(query14)
```

### 15. What are the top 10 flight routes by total revenue generated?

```python
query15 = '''
SELECT departure_airport, arrival_airport, SUM(tf.amount) AS route_revenue
FROM flights f
JOIN ticket_flights tf ON f.flight_id = tf.flight_id
GROUP BY departure_airport, arrival_airport
ORDER BY route_revenue DESC
LIMIT 10;
'''
run_query(query15)
```

### 16. What are the booking patterns and spending behaviors of different passengers?


```python
query16 = '''
SELECT passenger_id, COUNT(*) AS flight_count, SUM(amount) AS total_spent
FROM tickets t
JOIN ticket_flights tf ON t.ticket_no = tf.ticket_no
GROUP BY passenger_id
ORDER BY total_spent DESC
LIMIT 10;
'''
run_query(query16)
```
